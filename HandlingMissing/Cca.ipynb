```python
import pandas as pd    
import numpy as np      
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("dsa_job.csv")
df.head()
```

```python
df.isnull().mean()*100  # will show how much % data missing in each column
```

```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
df[cols]
```

```python
# if we remove 5% from above column let see how much data will remain
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50, density=True, figsize=(12, 12))
plt.show()
```
