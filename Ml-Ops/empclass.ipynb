```python
class emp:
    #A constructor in Python is a special method that is automatically called when you create a new object from a class.
    # It is used to initialize (set up) the object's data.
    def __init__(self):  
        self.id= 123
        self.name="ramrekha"
        self.age=26
        self.__salary = 50000   # private variable
    
    def travel(self,destination):
        print(f"i am now travellling to {destination}")
        print(self.age)
        print(self.__salary)


e1=emp()
print(e1.name)
```
