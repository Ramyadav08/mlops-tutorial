```python
class Bankaccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
            print(f"New balance: {self.balance}")
        else:
            print("Deposit amount must be positive")
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew: {amount}")
            print(f"New balance: {self.balance}")
        else:
            print("Insufficient funds or invalid withdrawal amount")
    
    def get_balance(self):
        return self.balance

Ram_account = Bankaccount("Ram", 1000)
Ram_account.deposit(500)
Ram_account.withdraw(200)
print(f"Final balance: {Ram_account.get_balance()}")
```

```python
# inheritance example
class Vehicle:
    def __init__(self, windows, door, engintype):
        self.windows = windows
        self.door = door
        self.engintype = engintype

    def drive(self):
        print(f"the person will drive the {self.engintype} vehicle")

class Car(Vehicle):
    def __init__(self, windows, door, engintype, brand):
        super().__init__(windows, door, engintype)
        self.brand = brand
    def selfdrive(self):
        print(f"the engine type is {self.engintype}")
        print(f"The {self.brand} car is self-driving car.")

tesla=Car(6, 4, "electric", "Tesla")
tesla.selfdrive()
tesla.drive()

        
```

```python
#multiple inheritance example
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} speaks")

class pet:
    def __init__(self, owner):
        self.owner = owner
    def play(self):
        print(f"{self.owner}'s pet is playing")
class Dog(Animal, pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        pet.__init__(self, owner)
    def speak(self):
        print("Woof! Woof!")
        super().speak()

dog=Dog("Buddy", "Alice")
dog.play()
dog.speak()
```

```python
#polymorphism example
class human:
    def __init__(self, name):
        self.name = name
    def work(self):  # method to be overloaded so we can not call it directly this is lost by the overloaded method
        print(f"{self.name} is working in an office.")
    
    def work(self,hours): # method overloading by changing number of parameters
        print(f"{self.name} is working for {hours} hours.")


class Engineer(human):
    def work(self,hours): # method overriding
        print(f"{self.name} is engineering for {hours} hours.")
        super().work(hours)


eng=Engineer("John")
eng.work(10)


```

```python
# encapsulation example
class Person:
    def __init__(self, name, age, mark):
        self.__name = name  # private attribute
        self.__age = age    # private attribute
        self._mark = mark    # protected attribute


    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Age cannot be negative")

class Person2(Person):
    def display_mark(self):
        print(f"Mark: {self._mark}")
        print(f"Name: {self.get_name()}")
        print(f"Age: {self.get_age()}")

person = Person("Alice", 30, 85)
print(person.get_name())
print(person.get_age())
person1 = Person2("Bob", 25, 90)
person1.display_mark()
# print(person.__age) # This will raise an AttributeError because __age is private
```

```python
# abstraction example
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        return 

    @abstractmethod
    def perimeter(self):
        pass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
rect = Rectangle(5, 10)
print(f"Area: {rect.area()}")
print(f"Perimeter: {rect.perimeter()}")
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius

circle = Circle(7)
print(f"Area: {circle.area()}")
print(f"Perimeter: {circle.perimeter()}")
```

```python
# decorators are powerful tools in Python that allow you to modify the behavior of functions or methods. Here's an example of a simple decorator that logs the execution time of a function:
# they are commonly used to add the functionality to functions or methods without modifying their actual code.
# decorator is just like function is passed as an argument to another function and extends the behavior of the latter function without explicitly modifying it.

def dcotr(fun):
    def wrapper():
        print("Before executing the function")
        fun()
        print("After executing the function")
    return wrapper


# def say_hello():
#     print("Hello!")

# decorated_say_hello = dcotr(say_hello)
# decorated_say_hello()

@dcotr
def say_hello():
    print("Hello!")
say_hello()



```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
@repeat(3)
def greet(name):
    print(f"Hello, {name}!")
greet("Alice")
```
