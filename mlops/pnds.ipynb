```python
import pandas as pd
#Series is one-dimensional array like object, it is similar to column in a table

data=[1,2,3,4,5]
sereis=pd.Series(data)
print(sereis)
```

```python
index=['a','b','c','d','e']
sereis=pd.Series(data,index=index)
print(sereis)
```

```python
data1={'a':1,'b':2,'c':3}
series_dict=pd.Series(data1)
print(series_dict)
#key consider as indes and values as series
```

```python
#dataframe a DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns). It is similar to a table in a relational database or an Excel spreadsheet.

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

print(df)
```
