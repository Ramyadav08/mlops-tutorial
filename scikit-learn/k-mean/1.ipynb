```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    "Customer":["ram","rekha","yadav","shyam","amit"],
    "Age":[23,45,31,35,40],
    "spending":[234,456,123,567,345]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
X=df[["Age","spending"]]
model=KMeans(n_clusters=2,random_state=42,n_init=10)

# random_state is set for reproducibility which means every time you run the code you will get the same result 
# which is important for debugging and comparing results.
#reproducibility means that the results of an experiment or analysis can be consistently replicated under the same conditions.
# why 42 ? 42 is often used as a humorous or arbitrary choice in programming and data science, popularized by Douglas Adams' "The Hitchhiker's Guide to the Galaxy," where 42 is described as the "answer to the ultimate question of life, the universe, and everything."

# n_init specifies the number of times the k-means algorithm will be run with different centroid seeds. The final results will be the best output of n_init consecutive runs in terms of inertia.
#centroid seeds are the initial positions of the cluster centers before the algorithm starts iterating to find the optimal clusters.
```

```python
df['Group']=model.fit_predict(X)
plt.figure(figsize=(6,5))

for group in df['Group'].unique():
    clustered_data = df[df['Group'] == group]
    plt.scatter(clustered_data['Age'], clustered_data['spending'], label=f'Group {group}')
    
plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer Segmentation using K-Means Clustering')
plt.legend()
plt.grid()
plt.show()
print("\nData with Cluster Groups:")
print(df)
```
