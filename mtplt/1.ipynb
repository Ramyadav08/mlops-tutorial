```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
y=[20, 60, 25, 70, 40]
# marker='o' specifies circular markers at each data point
# linestyle='--' creates a dashed line connecting the markers
# color='b' sets the line color to blue
#label='Data Line' adds a label for the legend
plt.plot(x, y, marker='o', linestyle='--', color='b', linewidth=2, markersize=6, label='Data Line')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Line Plot')   
plt.legend()  # Display the legend
plt.legend(loc='upper left')  # Position the legend in the upper left corner
plt.grid(linewidth=1, linestyle=':', color='gray')  # Add a grid with specified line width, style, and color
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
y=[20, 60, 25, 70, 40]
plt.plot(x, y, marker='o', linestyle='--', color='b', linewidth=2, markersize=6, label='Data Line')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sample Line Plot')   
plt.legend()  # Display the legend
plt.legend(loc='upper left')  # Position the legend in the upper left corner
plt.grid(linewidth=1, linestyle=':', color='gray')  # Add a grid with specified line width, style, and color
plt.xlim(1,4)  # Set x-axis limits from 1 to 4
plt.ylim(20,70)  # Set y-axis limits from 20 to 70
plt.show()
```

```python
#bar chart
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 60, 25, 70, 40]
plt.bar(categories, values, color='skyblue', edgecolor='black',label='categories values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Sample Bar Chart')
plt.legend()  # Position the legend in the upper left corner
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 60, 25, 70, 40]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sample Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
data = [20, 60, 25, 70, 40, 30, 50, 80, 90, 10]
plt.hist(data, bins=5, color='lightgreen', edgecolor='black')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')
plt.title('Sample Histogram')
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
hours_studied = [1, 2, 3, 4, 5]
scores = [50, 60, 70, 80, 90]
plt.scatter(hours_studied, scores, color='r', marker='^', label='Data Points')
plt.legend()  # Display the legend
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Sample Scatter Plot')
plt.grid()  # Add a grid for better visibility
plt.show()
```

```python
# scatter plot with multiple series
import matplotlib.pyplot as plt
hours_studied = [1, 2, 3, 4, 5]
scores_math = [50, 60, 70, 80, 90]
scores_science = [55, 65, 75, 85, 95]
plt.scatter(hours_studied, scores_math, color='r', marker='^', label='Math Scores')
plt.scatter(hours_studied, scores_science, color='b', marker='o', label='Science Scores')
plt.legend()  # Display the legend
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Sample Scatter Plot with Multiple Series')
plt.grid()  # Add a grid for better visibility
plt.show()
```

```python
#subplots
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [20, 60, 25, 70, 40]
y2 = [30, 50, 40, 60, 80]
fig, axs = plt.subplots(2, 1)  # Create a figure with 2 subplots (2 rows, 1 column) 
axs[0].plot(x, y1, marker='o', linestyle='--', color='b', label='Data Line 1')
axs[0].set_title('Subplot 1')
axs[0].legend()  # Display the legend for subplot 1
axs[1].plot(x, y2, marker='s', linestyle='-', color='r', label='Data Line 2')
axs[1].set_title('Subplot 2')
axs[1].legend()  # Display the legend for subplot 2
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
